```python
import pandas as pd
```

```python
df = pd.read_csv('res.csv', index_col=False)
df
```

```python
# drop last dates columns with uncomplete data
df = df.drop(labels=['2019-02-15', '2019-02-16', '2019-02-17'], axis=1)
df
```

```python
# change columns position
cols = df.columns.tolist()
cols_dates = cols[:-4]
cols_desc = ['Name of Card', 'Set', 'Foil or Not', 'Rarity']
cols = cols_desc + cols_dates

df_output = df[cols]
df_output
```

```python
df_output.to_csv('res.csv', index=False)
```
